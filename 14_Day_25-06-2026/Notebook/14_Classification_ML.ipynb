## Logistic Regression

### MODEL 1  -> Student Marks

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix , accuracy_score, precision_score, recall_score, f1_score



import warnings 
warnings.filterwarnings('ignore')

# data frame
data = {
"Marks": [35,45,55,65,75],
"StudyHours" : [1,2,3,4,5],
"Attendance" : [50,60,70,80,90],
"Result":["Fail", "Fail", "Pass", "Pass", "Pass"]
}
df = pd.DataFrame(data)



X = df[['Marks', 'StudyHours', 'Attendance']]
y = df['Result']

model = LogisticRegression()

model.fit(X,y)
print(df)
marks = float(input("Enter Marks :"))
hrs = float(input("Enter Ghante of Study :"))
attendence = float(input("Enter Attendence :"))

result = model.predict([[marks , hrs , attendence]])[0]

print("Prediciton :"  , result)


prob = model.predict_proba([[marks, hrs , attendence]])


print("\n Probability :")
print("Fail Probability :", round(prob[0][0] *100 ,2) ,"%")
print("Pass Probability :", round(prob[0][1] *100 ,2) ,"%")

cm = confusion_matrix(y, model.predict(X))
print("\n Confusion Matrix :")
print(cm)
TN, FP, FN, TP = cm.ravel()

print("\nTN =", TN)
print("FP =", FP)
print("FN =", FN)
print("TP =", TP)
# Manual Calculations
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)

print("\nManual Calculation")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)



y_pred = model.predict(X)
print("\nUsing sklearn")
print("Accuracy :", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred, pos_label='Pass'))
print("Recall   :", recall_score(y, y_pred, pos_label='Pass'))
print("F1 Score :", f1_score(y, y_pred, pos_label='Pass'))

# Create a scatter plot with a line of best fit
sns.regplot(x='StudyHours', y='Marks', data=df, 
            scatter_kws={'s': 100, 'color': 'blue'}, 
            line_kws={'color': 'red', 'linewidth': 2})

# Adding titles and labels
plt.title('Study Hours vs Marks (with Best Fit Line)', fontsize=14)
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Marks', fontsize=12)

# Display the plot
plt.show()
```

### MODEL 2 -> INSURACNCE PREDICTION

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
import warnings 
warnings.filterwarnings('ignore')

# data frame
data = {
    "Income": [20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000],
    "CreditScore": [500, 550, 580, 600, 620, 650, 680, 700, 720, 750],
    "LoanAmount": [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000],
    "Status": ["Rejected", "Rejected", "Rejected", "Rejected", "Approved", "Approved", "Approved", "Approved", "Approved", "Approved"]
}
df = pd.DataFrame(data)



X = df[['Income', 'CreditScore', 'LoanAmount']]
y = df['Status']

model = LogisticRegression()

model.fit(X,y)
print(df)
Income = float(input("Enter Income :"))
CreditScore = int(input("Enter CreditScore :"))
LoanAmount = float(input("Enter LoanAmount :"))

cm = confusion_matrix(y, model.predict(X))
print("\n Confusion Matrix :")  
print(cm)
result = model.predict([[Income , CreditScore , LoanAmount]])[0]

print("Prediciton :"  , result)

prob = model.predict_proba([[Income, CreditScore , LoanAmount]])


print("\n Probability :")
print("Fail Probability :", round(prob[0][0] *100 ,2) ,"%")
print("Pass Probability :", round(prob[0][1] *100 ,2) ,"%")

y_pred = model.predict(X)
accuracy = accuracy_score(y , y_pred)

print("Accuracy :" , accuracy*100 , "%")

# Create a scatter plot with a line of best fit
sns.regplot(x='CreditScore', y='Income', data=df, 
            scatter_kws={'s': 100, 'color': 'blue'}, 
            line_kws={'color': 'red', 'linewidth': 2})

# Adding titles and labels
plt.title('CreditScore vs Status (with Best Fit Line)', fontsize=14)
plt.xlabel('CreditScore', fontsize=12)
plt.ylabel('Income', fontsize=12)
# Display the plot
plt.show()
```
