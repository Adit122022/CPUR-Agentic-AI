## 1. Direct API vs LangChain
Using Direct Google GenAI API vs LangChain

```python
from google import genai

# Direct API Call
try:
    client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents="Explain Artificial Intelligence in one sentence."
    )
    print("Direct API Response:")
    print(response.text)
except Exception as e:
    print("Error:", e)
```

## 2. Prompt Templates & Chains

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

try:
    # Initialize LLM
    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0.7
    )

    # Create Prompt
    prompt = PromptTemplate.from_template("Explain {topic} in simple words.")

    # Create Chain
    chain = prompt | llm

    # Invoke Chain
    response = chain.invoke({"topic": "Python"})
    print("LangChain Response:")
    print(response.content)
except Exception as e:
    print("Error:", e)
```

## 3. Memory
Adding memory to keep track of conversation context.

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()

# Manually saving context
memory.save_context({"input": "My name is Asif"}, {"output": "Nice to meet you."})
memory.save_context({"input": "I am a Data Scientist"}, {"output": "That's great!"})

print("Memory Variables:")
print(memory.load_memory_variables({}))
```

## 4. RAG (Retrieval-Augmented Generation)
Reading from an external file and chunking it.

```python
# First, create a sample text file
with open("notes.txt", "w") as f:
    f.write("""Python is a high-level programming language.\nIt is used for AI, Machine Learning, Data Science, and Web Development.\nLangChain helps build LLM applications.""")

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

# Load Document
loader = TextLoader("notes.txt")
documents = loader.load()

# Split Document
splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=0)
chunks = splitter.split_documents(documents)

print("First Chunk Content:")
print(chunks[0].page_content)
```

## 5. Tools & Agents
Using external tools like DuckDuckGo Search and custom functions.

```python
from langchain.tools import tool

@tool
def calculator(expression: str):
    """Evaluates a mathematical expression."""
    return eval(expression)

print("Calculator Output for '10 * 20 + 5':")
print(calculator.invoke({"expression": "10 * 20 + 5"}))
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

# Initialize Search Tool
search = DuckDuckGoSearchRun()

# Search Internet
query = "What is LangChain?"
print(f"Searching for: {query}")
result = search.invoke(query)
print("Search Result:\n", result)
```

## 6. Chat Models & Messages
Understanding HumanMessage, AIMessage, SystemMessage.

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="You are a funny comedian."),
    HumanMessage(content="Tell me a joke about programming.")
]

try:
    response = llm.invoke(messages)
    print("Response with Messages:")
    print(response.content)
except Exception as e:
    print("Error:", e)
```

## 7. Chat Model Methods: stream() and batch()

```python
try:
    print("--- STREAMING RESPONSE ---")
    for chunk in llm.stream("Explain Machine Learning in 1 sentence."):
        print(chunk.content, end="")
    print("\n")
except Exception as e:
    print("Error:", e)
```

```python
try:
    print("--- BATCH RESPONSE ---")
    responses = llm.batch([
        "What is Python?",
        "What is AI?",
    ])

    for i, res in enumerate(responses):
        print(f"Response {i+1}:", res.content)
except Exception as e:
    print("Error:", e)
```
