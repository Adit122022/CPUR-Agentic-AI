```python
#deepCopy 
import copy
l1 = [10,20,[40,50]]
l2 = copy.deepcopy(l1)
# print(id(l1) , "  " , id(l2))
# Nested list is same
print(id(l1[2]) , "  " , id(l2[2]))   #2156974221440    2156982559232 

l2.append(30)
print(l1 , l2 , sep="\n")
```

```python
# After Lunch 
t = (20,40,80,10,50)
# Collection Function (Sabke liye work karega )
# len()
print("Tuple length :",len(t))

#max() 
print("Maximum in Collection  :" , max(t))

# min
print("Minimum in Collection  :" , min(t))

# sum()
print("Sum of elements :" ,sum(t))

# sorted 
print("Sorted Tuple (collection) :" , sorted(t))  # tuple -> converted into list

# reversed()
r = reversed(t)
print("Reverse of collection :"  , tuple(r))  # need to convert it into tuple() , list() , 

# enumerate()
print("\n \n =========Enumerate function============")
for i,v in enumerate(t):
    print(i, "," , v , end ="   |   ")

# zip 
print("\n \n =========ZIP function============", end="\n")
t = (20,40,80,10,50)
li=[10,40,20,80]
stri ="HELLO"
for l,tu in zip(stri,t):
    print(l , " -> " , tu , end = "  |  ")
```
