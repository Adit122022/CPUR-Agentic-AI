```python
# copy() 
# why do we need copying ?

# Deep Copy   , Shallow Copy

# Here the problem
# l1 = [10,20]
# l2 = l1
# print(id(l1) , "  " , id(l2))  #2156959520640    2156959520640
# l2.append(30)
# print(l1 , l2 , sep="\n")  
# [10, 20, 30]
# [10, 20, 30]


# Shallow Copy  -> created new list but nested list is the same

# l1 = [10,20]
# l2 = l1.copy()
# print(id(l1) , "  " , id(l2))
# l2.append(30)
# print(l1 , l2 , sep="\n")

#  Nexted list
l1 = [10,20,[40,50]]
l2 = l1.copy()
print(id(l1) , "  " , id(l2))
# Nested list is same
# print(id(l1[2]) , "  " , id(l2[2]))  #2156976268992    2156976268992

l2.append(30)
print(l1 , l2 , sep="\n")
```

```python
#deepCopy 
import copy
l1 = [10,20,[40,50]]
l2 = copy.deepcopy(l1)
# print(id(l1) , "  " , id(l2))
# Nested list is same
print(id(l1[2]) , "  " , id(l2[2]))   #2156974221440    2156982559232

l2.append(30)
print(l1 , l2 , sep="\n")
```
