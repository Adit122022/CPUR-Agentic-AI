## 🚀 Part 1: Deep Learning Training Process

Deep learning model training ek cycle hai jisme 3 key phases hote hain:
1. **Forward Propagation:** Input values weights aur biases se multiply hokar network se aage pass hoti hain aur final prediction $\hat{y}$ banti hai.
2. **Loss & Cost Calculation:** Hum prediction error check karte hain.
   - **Loss Function:** Ek single example ke error ko calculate karta hai (e.g. Binary Cross Entropy or Mean Squared Error).
   - **Cost Function:** Pure dataset (ya batch) ke losses ka average calculated value hota hai.
3. **Backpropagation & Gradient Descent:** Calculus ke chain rule ko use karke error derivative find karte hain aur weights ko update karte hain ($W = W - \eta \cdot \text{Gradient}$).

## 🔢 Part 2: Handwritten Digit Recognition using TensorFlow/Keras

Hum **MNIST** dataset ka use karenge jisme handwritten digits (0-9) ki 70,000 images hain.

### Step 1: Import Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Flatten, Dense
```

### Step 2: Load Dataset

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Training Images Shape :", X_train.shape)
print("Training Labels Shape :", y_train.shape)
print("Testing Images Shape :", X_test.shape)
print("Testing Labels Shape :", y_test.shape)
```

### Step 3: Display Sample Images

```python
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i], cmap="gray")
    plt.title(f"Label : {y_train[i]}")
    plt.axis("off")
plt.tight_layout()
plt.show()
```

### Step 4: Normalize Images

Grayscale pixel values 0-255 ki range mein hoti hain. Model training ko stable aur fast karne ke liye hum pixel values ko 255.0 se divide karke 0-1 scaling normalize karte hain.

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

### Step 5: Build Neural Network

Hum model sequential layer design karenge:
- **Flatten Layer:** $28 \times 28$ image shape ko flat karke 784 pixels array banati hai.
- **Dense Layers:** Features seekhne ke liye ReLU activations layers.
- **Output Dense:** 10 categories ki probability distribution nikalne ke liye Softmax activation function layer.

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

### Step 6: Model Summary

```python
model.summary()
```

### Step 7: Compile Model

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

### Step 8: Train Model

```python
history = model.fit(
    X_train, 
    y_train, 
    epochs=10, 
    batch_size=32, 
    validation_split=0.2
)
```

### Step 9: Evaluate Model on Test Data

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\nTest Loss : {loss:.4f}")
print(f"Test Accuracy : {accuracy*100:.2f}%")
```

### Step 10 & 11: Plot Accuracy & Loss Curves

```python
plt.figure(figsize=(12, 4))

# Accuracy curve
plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

# Loss curve
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

### Step 12: Predict Single Image

```python
prediction = model.predict(X_test[0:1])
predicted_digit = np.argmax(prediction)

print("\nPredicted Digit :", predicted_digit)
print("Actual Digit    :", y_test[0])

plt.imshow(X_test[0], cmap="gray")
plt.title(f"Prediction : {predicted_digit}")
plt.axis("off")
plt.show()
```

### Step 13: Predict Multiple Images

```python
plt.figure(figsize=(12,8))
for i in range(12):
    prediction = model.predict(X_test[i:i+1], verbose=0)
    digit = np.argmax(prediction)
    plt.subplot(3,4,i+1)
    plt.imshow(X_test[i], cmap="gray")
    plt.title(f"P:{digit}  A:{y_test[i]}")
    plt.axis("off")
plt.tight_layout()
plt.show()
```

### Step 14: Save Model

```python
model.save("digit_recognition.keras")
print("\nModel Saved Successfully!")
```

### Step 15 & 16: Load Saved Model & Test

```python
new_model = load_model("digit_recognition.keras")
print("Model Loaded Successfully!")

prediction = new_model.predict(X_test[25:26])
digit = np.argmax(prediction)

print("\nPrediction from Loaded Model :", digit)
print("Actual Digit                 :", y_test[25])
plt.imshow(X_test[25], cmap="gray")
plt.title(f"Loaded Model Prediction : {digit}")
plt.axis("off")
plt.show()
```

### Step 17: Predict Random Images

```python
print("\nRandom Predictions\n")
for i in np.random.randint(0, len(X_test), 5):
    prediction = new_model.predict(X_test[i:i+1], verbose=0)
    digit = np.argmax(prediction)
    print("----------------------------")
    print("Image Index :", i)
    print("Predicted   :", digit)
    print("Actual      :", y_test[i])
    plt.imshow(X_test[i], cmap="gray")
    plt.title(f"Predicted : {digit}")
    plt.axis("off")
    plt.show()
print("\nProject Completed Successfully!")
```
