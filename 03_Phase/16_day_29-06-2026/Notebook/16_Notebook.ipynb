````python
# What is an API?
'''
**API (Application Programming Interface)** is a bridge that allows
two software applications to communicate with each other.

Think of an API as a **messenger**.
* You (Client)
* Waiter (API)
* Kitchen (Server)
* Food (Response)
The waiter takes your order to the kitchen and brings your food back.

Exactly the same happens in software.
User
   │
   ▼
Application
   │
   ▼
API
   │
   ▼
Server
   │
   ▼
Database / AI Model


# Real-Life Examples of APIs

## Example 1: Weather App
You open a weather app.
Weather App
      │
      ▼
Weather API
      │
      ▼
Weather Database
Response
Temperature : 32°C
Humidity : 60%
Wind : 10 km/hr

## Example 2: Google Maps
Uber doesn't create maps.
Instead it uses
Google Maps API
to get
* Roads
* Distance
* Directions
* Traffic
'''
````

````python
# Why APIs are Important?
'''
APIs (**Application Programming Interfaces**) are essential because they allow different
software applications to communicate and share data or services without needing to know
each other's internal implementation.

# Why Do We Need APIs?
Imagine every application had to build every feature from scratch.
For example:
* A shopping website would need to create its own payment system.
* A taxi app would need to build its own maps.
* A weather app would need to install weather stations worldwide.
This would be expensive, time-consuming, and inefficient.


# Importance of APIs

## 1. Easy Communication Between Applications
APIs allow two different applications to exchange information.
### Example
Swiggy
     │
     ▼
Payment API
     │
     ▼
Bank Server
Swiggy sends payment information to the bank using an API and receives a success or failure response.

## 2. Save Development Time
Instead of creating every feature yourself, you can use existing APIs.
### Example
Without API
* Build your own payment system
* Build your own map service
* Build your own email service

With API
* Razorpay API
* Google Maps API
* Gmail API
This reduces development time from months to days.

## 3. Reuse Existing Services
APIs let developers reuse powerful services created by other companies.
### Examples
| Service  | API Used        |
| -------- | --------------- |
| Maps     | Google Maps API |
| Payments | Razorpay API    |
| AI       | OpenAI API      |
| Email    | Gmail API       |
| SMS      | Twilio API      |
| Weather  | OpenWeather API |

## 4. Faster Application Development
Developers can focus on building unique features instead of reinventing common services.
### Example
An AI Resume Builder only needs to:
* Collect user details
* Send a prompt to an AI API
* Display the generated resume
The AI model does the complex work.

## 5. Secure Data Sharing
APIs use authentication methods such as:
* API Keys
* OAuth
* Bearer Tokens
* JWT Tokens
Only authorized users and applications can access the service.

## 6. Cloud-Based Access
Modern APIs work over the internet.
You don't need to install large software locally.

## 7. Scalability
APIs allow applications to handle millions of users efficiently.
### Example
Instagram
Millions of users upload photos every day because APIs and cloud infrastructure can scale as demand increases.

## 8. Automation
APIs automate repetitive tasks

## 9. AI Integration
Today, AI capabilities are added to applications through APIs.
Examples:
* AI Chatbots
* AI Email Writers
* AI Resume Builders
* AI Code Assistants
* AI Translators
* AI Content Generators

## 10. Platform Independence
Applications built using different programming languages can communicate through APIs.
Example:
* Frontend → React
* Backend → Python
* Database → MySQL
They exchange data using APIs.

'''
````

````python
# Types of APIs
'''
An **API (Application Programming Interface)** enables different software applications
to communicate with each other. APIs can be classified in two main ways:
1. **Based on Architecture (How the API works)**
2. **Based on Access (Who can use the API)**

# Part 1: Types of APIs Based on Architecture
There are four major architectural styles:
1. REST API
2. SOAP API
3. GraphQL API
4. gRPC API

# 1. REST API (Representational State Transfer)
REST API is the **most popular and widely used** type of API for web and mobile applications.
It communicates using the **HTTP protocol** and exchanges data primarily in **JSON** format.
## Architecture
Client
   │
HTTP Request
   │
REST API
   │
Server
   │
Database
   │
HTTP Response (JSON)
   │
Client

## Features
* Uses HTTP protocol
* Fast and lightweight
* Easy to understand
* Platform independent
* Uses JSON format
* Stateless (each request is independent)


## HTTP Methods
### GET
Retrieve data
GET /students

### POST
Create data
POST /students

### PUT
Update complete data
PUT /students/5

### PATCH
Update partial data
PATCH /students/5

### DELETE
Delete data
DELETE /students/5

## Real-World Examples
* OpenAI API
* GitHub API
* Spotify API
* Google Maps API
* Weather API
* Twitter (X) API

# 2. SOAP API (Simple Object Access Protocol)
SOAP is an older API protocol used in enterprise applications
where **security, reliability, and transactions** are critical.

SOAP uses **XML** instead of JSON.
## Architecture
Client
   │
SOAP XML Request
   │
SOAP Server
   │
Database
   │
SOAP XML Response
   │
Client

## Features
* Uses XML only
* Very secure
* Strict standards
* Built-in error handling
* Supports transactions

## Example XML
<student>
    <name>Asif</name>
    <course>Python</course>
</student>
## Where SOAP is Used
* Banking
* Insurance
* Healthcare
* Government
* Financial Systems

# 3. GraphQL API
GraphQL is a modern API technology developed by **Meta (Facebook)**.
It allows the client to request **only the data it needs**,
reducing unnecessary data transfer.

## Architecture
Client
   │
GraphQL Query
   │
GraphQL Server
   │
Database
   │
Requested Data Only


## Example Query
{
  student {
    name
    course
  }
}


## Response
{
  "student": {
    "name": "Asif",
    "course": "Python"
  }
}


## Features
* Flexible
* Fast
* Fetch only required fields
* Single endpoint
* Reduces network traffic

## Used By
* Facebook
* Shopify
* GitHub
* Netflix (in some services)


# 4. gRPC API
gRPC is a high-performance API framework developed by **Google**.
It uses **Protocol Buffers (Protobuf)** instead of JSON, making it extremely fast.

## Architecture
Client
   │
Protocol Buffers
   │
gRPC Server
   │
Database

## Features
* Extremely fast
* Small data size
* Supports streaming
* Cross-platform
* Strongly typed

## Used By
* Google
* Netflix
* Uber
* Cloud Services
* Microservices

# Part 2: Types of APIs Based on Access
# 1. Public API (Open API)
Anyone can use it after registration or obtaining an API key.

### Examples
* OpenWeather API
* NASA API
* GitHub API

# 2. Private API
Used **inside an organization** only.
### Example
A company's HR system communicating with its Payroll system.

# 3. Partner API
Shared only with **trusted business partners**.
### Example
An airline sharing booking APIs with travel websites.

# 4. Composite API
Combines multiple API calls into **one request**, reducing the number of network calls.
### Example
An e-commerce app fetching:
* User profile
* Order history
* Payment details

# Which API Should You Use?
| Use Case                    | Recommended API     |
| --------------------------- | ------------------- |
| Web Application             | REST API            |
| Mobile Application          | REST API or GraphQL |
| Banking & Financial Systems | SOAP API            |
| AI Applications             | REST API            |
| Microservices               | gRPC                |
| Real-Time Applications      | gRPC                |
| Large Enterprise Systems    | SOAP or gRPC        |
'''
````
