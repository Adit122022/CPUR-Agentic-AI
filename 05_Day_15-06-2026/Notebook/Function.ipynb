## Functions

A **function** is a reusable block of code that performs a specific task.
Functions help organize code, reduce repetition, and make programs easier to understand and maintain.

- Advantages of Functions
- Code Reusability: Write once and use multiple times.
- Reduces Code Duplication: Avoids writing the same code repeatedly.
- Improves Readability: Programs become easier to understand.
- Easy Maintenance: Changes need to be made in only one place.
- Better Testing & Debugging: Functions can be tested independently.
- Modular Programming: Large programs can be divided into smaller modules.

# Recursive Function

```python
# Create a fucntion for calculator

add = lambda n1, n2 : n1 + n2
sub = lambda n1,n2 : n1 - n2
mul = lambda n1,n2 : n1 * n2
div = lambda n1,n2 : n1 // n2
rem = lambda n1,n2 : n1 % n2

# a, b =map(int,input("ENter two numbers with comma(,) seperated :").split(","))

def calculator():
    a, b =map(int,input("ENter two numbers with comma(,) seperated :").split(","))
    ch = input("Enter choice or symobol to calculate (+,-,*,/,%) \n if u want to end =0 ")
    if ch == "0":
        return "COME AGAIN !"
    match ch:
        case "+":
            print("Addition :",add(a,b))
        case "-":
            print("Subtraction :",sub(a,b))
        case "*":
            print("Multiplication :",mul(a,b))
        case "/":
            print("Division :",div(a,b))
        case "%":
            print("Remainder :",rem(a,b))
        case _:
            print("ERROR:INVALID NUMBER")
    calculator(a,b)
calculator(a,b)
```

- infinite recursive function is runs until the stack memorey fulls

```python
# Sum of numbers from 1 to n using recursion
def sum_to_n(n):
    if n == 0:
        return 0
    return n+ sum_to_n(n-1)
   
print(sum_to_n(10))
```

```python
# Sum of digits of a number using recursion 
def sum_of_digits(n):
    if n == 0:
        return 0
    return n%10 + sum_of_digits(n//10)

print(sum_of_digits(123))
```

```python
# Factorial of a number using recursion
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
print(factorial(5))
```

```python
# Generator Function

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
print("Counting up to 5:", tuple(count_up_to(5)))
```

# Scientific calculator

```python
import math

# ── Basic Operations (Lambda Functions) ─────────────────────────────────────
add   = lambda a, b: a + b
sub   = lambda a, b: a - b
mul   = lambda a, b: a * b
div   = lambda a, b: a / b   if b != 0 else "Error: Division by zero!"
mod   = lambda a, b: a % b   if b != 0 else "Error: Division by zero!"
power = lambda a, b: a ** b

# ── Scientific Operations (Lambda Functions) ─────────────────────────────────
square   = lambda n: n ** 2
cube     = lambda n: n ** 3
sqrt_f   = lambda n: math.sqrt(n)   if n >= 0 else "Error: Negative sqrt!"
inverse  = lambda n: 1 / n          if n != 0 else "Error: Division by zero!"
absolute = lambda n: abs(n)

# Trig — input in degrees (converted inside lambda)
sin_d = lambda n: round(math.sin(math.radians(n)), 10)
cos_d = lambda n: round(math.cos(math.radians(n)), 10)
tan_d = lambda n: round(math.tan(math.radians(n)), 10) \
                  if abs(math.cos(math.radians(n))) > 1e-10 else "Error: Undefined!"

log10_f = lambda n: math.log10(n) if n > 0 else "Error: Domain error!"
ln_f    = lambda n: math.log(n)   if n > 0 else "Error: Domain error!"

# Constants as lambdas
PI = lambda: math.pi
E  = lambda: math.e

# ── Recursive: Factorial ─────────────────────────────────────────────────────
def factorial(n):
    """Factorial using recursion — n!"""
    if not isinstance(n, int) or n < 0:
        return "Error: Enter a non-negative integer!"
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# ── Display Menu ─────────────────────────────────────────────────────────────
def show_menu():
    print("""
╔══════════════════════════════════════════╗
║        🧮  SCIENTIFIC CALCULATOR  🧮     ║
╠══════════════════════════════════════════╣
║  BASIC OPERATIONS                        ║
║   1. Addition       (+)                  ║
║   2. Subtraction    (-)                  ║
║   3. Multiplication (*)                  ║
║   4. Division       (/)                  ║
║   5. Modulo         (%)                  ║
║   6. Power          (a^b)                ║
╠══════════════════════════════════════════╣
║  SCIENTIFIC FUNCTIONS (single number)    ║
║   7. Square         (x²)                 ║
║   8. Cube           (x³)                 ║
║   9. Square Root    (√x)                 ║
║  10. Inverse        (1/x)                ║
║  11. Absolute Value (|x|)                ║
║  12. Factorial      (n!)                 ║
╠══════════════════════════════════════════╣
║  TRIGONOMETRY (degrees)                  ║
║  13. sin(x)                              ║
║  14. cos(x)                              ║
║  15. tan(x)                              ║
╠══════════════════════════════════════════╣
║  LOGARITHM                               ║
║  16. log₁₀(x)                            ║
║  17. ln(x)  — Natural log                ║
╠══════════════════════════════════════════╣
║  CONSTANTS                               ║
║  18. π  (Pi)                             ║
║  19. e  (Euler's number)                 ║
╠══════════════════════════════════════════╣
║   0. Exit                                ║
╚══════════════════════════════════════════╝
""")

# ── Input Helpers (Lambda) ────────────────────────────────────────────────────
get_two = lambda: tuple(map(float, input("  Enter two numbers (a b): ").split()))
get_one = lambda: float(input("  Enter a number: "))

# ── Result Formatter (Lambda) ─────────────────────────────────────────────────
fmt = lambda v: int(v) if isinstance(v, float) and v == int(v) else v

# ── Operation Dispatcher — dict of (label, lambda, num_args) ─────────────────
def get_op_map():
    """Returns a dict mapping choice → (label, function, num_args)"""
    return {
        '1' : ('Addition',       lambda a, b: add(a, b),       2),
        '2' : ('Subtraction',    lambda a, b: sub(a, b),       2),
        '3' : ('Multiplication', lambda a, b: mul(a, b),       2),
        '4' : ('Division',       lambda a, b: div(a, b),       2),
        '5' : ('Modulo',         lambda a, b: mod(a, b),       2),
        '6' : ('Power',          lambda a, b: power(a, b),     2),
        '7' : ('Square',         lambda n, _: square(n),       1),
        '8' : ('Cube',           lambda n, _: cube(n),         1),
        '9' : ('Square Root',    lambda n, _: sqrt_f(n),       1),
        '10': ('Inverse',        lambda n, _: inverse(n),      1),
        '11': ('Absolute Value', lambda n, _: absolute(n),     1),
        '12': ('Factorial',      lambda n, _: factorial(int(n)), 1),
        '13': ('sin',            lambda n, _: sin_d(n),        1),
        '14': ('cos',            lambda n, _: cos_d(n),        1),
        '15': ('tan',            lambda n, _: tan_d(n),        1),
        '16': ('log₁₀',          lambda n, _: log10_f(n),      1),
        '17': ('ln',             lambda n, _: ln_f(n),         1),
        '18': ('π  (Pi)',        lambda _, __: PI(),           0),
        '19': ('e  (Euler)',     lambda _, __: E(),            0),
    }

# ── Main Calculator Function ──────────────────────────────────────────────────
def scientific_calculator():
    """Scientific Calculator using def + lambda functions only."""
    op_map = get_op_map()

    while True:
        show_menu()
        ch = input("  Enter your choice (0 to exit): ").strip()

        if ch == '0':
            print("\n  👋 Bye bye! Calculator band ho gaya. Come Again!")
            break

        if ch not in op_map:
            print("  ❌ Invalid choice! Please choose from the menu.\n")
            continue

        label, fn, num_args = op_map[ch]

        try:
            if num_args == 2:
                a, b = get_two()
                result = fn(a, b)
            elif num_args == 1:
                a = get_one()
                result = fn(a, None)
            else:                        # constants
                result = fn(None, None)

            print(f"\n  ✅  {label}  →  {fmt(result)}\n")

        except ValueError:
            print("  ❌ Invalid input! Please enter numeric values.\n")
        except Exception as ex:
            print(f"  ❌ Error: {ex}\n")

# ── Run It! ───────────────────────────────────────────────────────────────────
scientific_calculator()
```
