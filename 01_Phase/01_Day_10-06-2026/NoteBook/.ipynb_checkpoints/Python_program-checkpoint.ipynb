# Project Objective

## Create a Python program that asks for a student's name and marks, calculates the total and
percentage, and displays the result.
Requirements
• Use print()
• Use input()
• Use variables
• Use arithmetic operators (+ and /)

- Sample Output
  
 MARKSHEET SYSTEM
Enter Student Name: Asif
Enter Python Marks: 80
Enter Maths Marks: 90
Enter English Marks: 70




- Result
Student Name : Asif
Total Marks : 240
Percentage : 80.0
Thank You!

```python
print("==============MARKSHEET SYSTEM=============" ,"STUDENT DETAILS -" , sep="\n")

name = input("Enter Student Name : ")
py_marks = float(input("ENter python marks :"))
math_marks = float(input("ENter maths marks :"))
english_marks = float(input("ENter English marks :"))
total = py_marks + math_marks + english_marks
p = round( total/ 3,2)

print(f"==============RESULT=============" ,f"Student Name :{name}",f"Total Marks obtained by Student :  {total}", f"Percentage :{p}" , sep="\n")
```
