```python
a = int(input("Number dalo :"))
a
```

### Taking Multiple Inputs
We can take multiple inputs in a single line using the `split()` function, which splits a string into a list.
The `map()` function is then used to apply a function (like `float`) to all the items in that list.


```python
a ,b = map(float , input("Enter Number :").split(" "))
print(a+b)
```

## 2. Variables & Data Types
Variables are containers for storing data values. Python has various standard data types:
- **Numbers**: `int` (integers), `float` (decimals), `complex` (complex numbers)

- **Boolean**: `bool` (`True` or `False`)


Let's look at an example where we take a user's name, roll number, and age, and print them formatted.

## 3. Arithmetic Operators
Arithmetic operators are used with numeric values to perform common mathematical operations:
- `+` : Addition
- `-` : Subtraction
- `*` : Multiplication
- `/` : Division
- `//`: Floor Division (rounds down to the nearest whole number)
- `%` : Modulus (returns the remainder)
- `**`: Exponentiation (power)


```python

a, b = map(int, input("Enter numbers (separated with comma): ").split(","))

print("Addition :", a + b)          
print("Subtraction :", a - b)       
print("Multiply :", a * b)          
print("Division :", a / b)          
print("Floor Division :", a // b)   
print("Modulus (Remainder) :", a % b) 
print("Power (Exponent) :", a ** b) # 2 ** 2 ** 3 isme phele  2 **( 2 **3)
```

## 4. Relational (Comparison) Operators
Relational operators compare two values and return a Boolean result (`True` or `False`).
- `==` : Equal
- `!=` : Not equal
- `>`  : Greater than
- `<`  : Less than
- `>=` : Greater than or equal to
- `<=` : Less than or equal to


```python
a = 10
b = 5
print(a == b)
print(a!=b)
print( a < b )
print( a > b )
print( a >= b )
print( a <= b )

```

## 5. Logical Operators
Logical operators are used to combine conditional statements:
- `and`: Returns `True` if both statements are true
- `or` : Returns `True` if one of the statements is true
- `not`: Reverse the result, returns `False` if the result is true


```python
print("ravi is awesome") and print("everyone is awesome")
```
