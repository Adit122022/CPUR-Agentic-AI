```python
l_name ="Arya College Library"
location = "Jaipur"

b_stock = {
    "Python Programming": 10,
    "Data Structures": 8,
    "DBMS": 6,
    "Computer Networks": 5,
    "Operating System": 4
}
categories = [
    "Programming",
    "Database",
    "Networking",
    "System"
]

print("="*40, "    College Library System","="*40, sep="\n")
print("\n",f"Library Name : {l_name}" , f"Library Location : {location}" , sep="\n")

print("\nBook Categories :  ",end=" ")
for c in categories:
    print(c  , end = "  |  ")

    
print("\n\nBook Available in Library  : ")
for i, b in enumerate(b_stock.keys(), 1):
    print(f"{i}. {b}")

print("\nBook Stock : ")
for b, s in b_stock.items():
  
    print(f"{b:<19}: {s}")

print("\nChecking Book Availability...\n")
s_book = input("Enter Book Name : ")

if s_book in b_stock:
    print("Book Available in Library")
    
else:
    print("Book Not Available in Library")


print(f"Total Books Titles : {len(b_stock)}")
print(f"Total Categories   : {len(categories)}")
```
