# 🎲 SET — "No Duplicates Club!"
- **Unordered** → Elements ka koi fixed order nahi
- **Mutable** → Add/Remove kar sakte ho
- **Unique Values** → Duplicates auto-remove ho jaate hain 🧹
- **Not Indexed** → `s[0]` nahi chalega ❌

### Set Methods:
| Method | Kya karta hai |
|--------|---------------|
| `add(val)` | Ek element add karo |
| `update([...])` | Multiple elements add karo |
| `remove(val)` | Hatao — nahi mila toh **ERROR** 💥 |
| `discard(val)` | Hatao — nahi mila toh **kuch nahi** 😌 |
| `pop()` | Random ek element nikalo 🎲 |
| `clear()` | Sab saaf ☢️ |
| `union(s2)` / `\|` | Dono sets mila do (A ∪ B) |
| `intersection(s2)` / `&` | Sirf common elements (A ∩ B) |
| `difference(s2)` / `-` | A mein hai, B mein nahi (A - B) |

```python
# ============================================================
#  SET — Creation & Properties
# ============================================================

s1 = {10, 20, 30, 10, 40}   # Duplicate 10 auto-remove!
print("Set:", s1)            # Order random hoga

# Empty set banana — {} mat karo! woh dict banta hai!
s_empty = set()
print("Type of {}  :", type({}))       # <class 'dict'>
print("Type of set():", type(s_empty))  # <class 'set'>

# add() — ek element
s1.add(70)
print("After add(70):", s1)

# update() — multiple elements
s1.update(["H", "E", "l", "L"])
print("After update:", s1)

# remove() vs discard()
s1.remove("H")      # 'H' nikaal do
print("After remove('H'):", s1)

s1.discard("H")     # 'H' already nahi hai — koi error nahi!
print("After discard('H') [no error]:", s1)
```

```python
# ============================================================
#  SET OPERATIONS — Union, Intersection, Difference
# ============================================================

s1 = {10, 20, 30}
s2 = {20, 30, 40, 50}

# Union — dono ke sab (unique)
print("Union (s1 | s2)        :", s1 | s2)             # {10, 20, 30, 40, 50}
print("Union (s1.union(s2))   :", s1.union(s2))        # Same

# Intersection — sirf common
print("Intersection (s1 & s2) :", s1 & s2)             # {20, 30}
print("Intersection (method)  :", s1.intersection(s2)) # Same

# Difference — s1 mein hai, s2 mein nahi
print("Difference (s1 - s2)   :", s1 - s2)             # {10}
print("Difference (method)    :", s1.difference(s2))   # Same
```

# ❄️ FROZEN SET — "Set Ka Immutable Bhai"
- **Immutable** → Ek baar bana, change nahi hoga
- **Unique Values** → Set ki tarah
- **Unordered** → Random order
- **Hashable** → Dict key ban sakta hai! (Normal set nahi ban sakta)

### Kya allowed hai?
| Operation | Allowed? |
|-----------|----------|
| `len()`, `in`, `not in` | ✅ Haan |
| `union`, `intersection`, `difference` | ✅ Haan |
| `add()`, `remove()`, `update()` | ❌ Nahi (Immutable!) |

```python
# ============================================================
#  FROZEN SET
# ============================================================

f1 = frozenset((10, 20, 30, 40, 50, "Hello"))  # Tuple se banao
f2 = frozenset([10, 20, 20, 30])               # List se (duplicates remove!)

print("f1:", f1)   # frozenset({10, 20, 30, 40, 50, 'Hello'})
print("f2:", f2)   # frozenset({10, 20, 30})

# Collection functions — chalte hain
print("Length:", len(f1))     # 6
print("10 in f1:", 10 in f1)  # True

# Set operations — chalte hain
print("Union:", f1 | f2)

# Modification — nahi chalega!
try:
    f1.add(99)
except AttributeError as e:
    print("Error:", e)  # 'frozenset' object has no attribute 'add'

# Frozen Set as Dictionary Key (normal set nahi ban sakta!)
my_dict = {f2: "Yeh frozenset key hai!"}
print("Dict with frozenset key:", my_dict)
```

# 📖 DICTIONARY — "Python Ka Google Maps 🗺️"
- **Key-Value Pairs** → Har cheez naam se milti hai
- **Mutable** → Change kar sakte ho
- **Ordered** → Python 3.7+ insertion order maintain karta hai
- **No Duplicate Keys** → Same key = purani value replace
- **O(1) Lookup** → Sabse fast search!

### Dictionary Methods:
| Method | Kya karta hai |
|--------|---------------|
| `keys()` | Saare keys deta hai 🔑 |
| `values()` | Saari values deta hai 💰 |
| `items()` | (key, value) pairs deta hai 📦 |
| `get(key)` | Safe access — KeyError nahi aayega ✅ |
| `pop(key)` | Key-value delete karo 💥 |
| `popitem()` | Last pair delete karo (LIFO) 🔄 |
| `update({...})` | Merge/update karo 🔄 |
| `clear()` | Sab saaf ☢️ |

```python
# ============================================================
#  DICTIONARY — Creation & Accessing Values
# ============================================================

d1 = {
    "naam":  "Aditya",
    "age":   20,
    "city":  "Kota"
}

# Mixed key types allowed!
d2 = {
    "Kota": (1, 2),
    2:       "AHHHHHHHHHH"
}

# Direct Access
print(d1["naam"])    # Aditya
print(d2["Kota"])   # (1, 2)

# get() — safe access (KeyError nahi aata)
print(d1.get("age"))           # 20
print(d1.get("phone"))         # None
print(d1.get("phone", "N/A"))  # N/A (default value)
```

```python
# ============================================================
#  DICTIONARY — keys(), values(), items(), Traversal
# ============================================================

d1 = {"naam": "Aditya", "age": 20, "city": "Kota"}

print("Keys   :", list(d1.keys()))    # ['naam', 'age', 'city']
print("Values :", list(d1.values())) # ['Aditya', 20, 'Kota']
print("Items  :", list(d1.items()))  # [('naam', 'Aditya'), ...]

print("\n--- Traversal using items() ---")
for k, v in d1.items():
    print(f"{k} → {v}")

# pop() — specific key
popped = d1.pop("city")
print("\nPopped value:", popped)  # Kota
print("After pop   :", d1)

# update() — naya data merge karo
d1.update({"phone": "9876543210", "age": 21})  # age bhi update!
print("After update:", d1)
```

```python
# ============================================================
#  NESTED DICTIONARY — "Dictionary Andar Dictionary! 🤯"
# ============================================================

students = {
    "Aditya": {"age": 20, "marks": 95},
    "Rohit":  {"age": 21, "marks": 88}
}

print(students["Aditya"])           # {'age': 20, 'marks': 95}
print(students["Aditya"]["marks"]) # 95
print(students["Rohit"]["age"])    # 21

# Traversal in nested dict
print("\n--- Student Report ---")
for name, info in students.items():
    print(f"{name}: Age={info['age']}, Marks={info['marks']}")
```
