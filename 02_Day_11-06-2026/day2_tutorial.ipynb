## 1. Dynamic Typing
In Python, you don't need to declare the type of a variable. The type is determined at runtime.

```python
x = 10
print(type(x))
x = "Hello"
print(type(x))
```

```python
num_str = "100"
num_int = int(num_str)
print(type(num_int))
```

## 5. Operators

### Arithmetic Operators
`+`, `-`, `*`, `/`, `//` (floor division), `%` (modulus), `**` (exponentiation)

### Relational Operators
`==`, `!=`, `>`, `<`, `>=`, `<=`

### Logical Operators
`and`, `or`, `not`

### Bitwise Operators
`&`, `|`, `^`, `~`, `<<`, `>>`

### Membership Operators
`in`, `not in`

### Identity Operators
`is`, `is not`

```python
a, b = 10, 3
print("Arithmetic:", a + b, a // b)
print("Relational:", a > b)
print("Logical:", a > 5 and b < 5)
print("Bitwise:", a & b)
print("Membership:", 3 in [1, 2, 3])
print("Identity:", a is b)
```

## 6. Operator Precedence Table
Python operators have an order of precedence. (See README for article link)
Example: Exponentiation `**` is evaluated before Multiplication `*`.

```python
print(10 + 2 * 3) # Multiplication happens first
```

## 7. Basic Collections

- **List:** Ordered, mutable (`[1, 2, 3]`)
- **Tuple:** Ordered, immutable (`(1, 2, 3)`)
- **Set:** Unordered, unique elements (`{1, 2, 3}`)
- **Frozenset:** Immutable set
- **Dict:** Key-value pairs (`{'a': 1}`)

```python
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_set = {1, 2, 3}
my_dict = {"name": "Alice", "age": 25}
my_frozenset = frozenset([1, 2, 3])

print(type(my_list), type(my_tuple), type(my_set), type(my_dict), type(my_frozenset))
```
