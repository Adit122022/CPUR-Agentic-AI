```python
e_name = input("Enter employee Name :")
e_id = int(input("Enter employee id :"))
salary = float(input("Enter employee salary :"))
e_status = bool(input("Enter employee status ( 0 for false 1 for true ) :"))


print(f"Employee Id :{e_id}",f"Employee name : {e_name}" ,f"Salary : {salary}" ,f"Employee Status :{str(e_status)}" , sep="\n")
print("Memory Location- ")
print(f"Employee Id :{id(e_id)}",f"Employee name : {id(e_name)}" ,f"Salary : {id(salary)}" ,f"Employee Status :{id(e_status)}" , sep="\n")
print("Types- ")
print(f"Employee Id :{type(e_id)}",f"Employee name : {type(e_name)}" ,f"Salary : {type(salary)}" ,f"Employee Status :{type(e_status)}" , sep="\n")
```
