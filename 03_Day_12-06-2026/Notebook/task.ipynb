```python


print("=" * 45)
print("         CRICKET SCORE ANALYZER")
print("=" * 45)

ovr = int(input("Enter Number of ovr Played : "))

t_run = 0
fours = 0
sixes = 0
d_ball = 0

for over in range(1, ovr + 1):

    print(f"\n---------- Over {over} ----------")

    for ball in range(1, 7):

        runs = int(input(f"Ball {ball} Runs (0-6) : "))

        # Validation
        if runs < 0 or runs > 6 or runs == 5:
            print("Invalid Runs! Counting as Dot Ball.")
            runs = 0

        t_run += runs

        if runs == 4:
            fours += 1

        elif runs == 6:
            sixes += 1

        elif runs == 0:
            d_ball += 1


t_ball = ovr * 6
boundry = (fours * 4) + (sixes * 6)
st_rate = (t_run / t_ball) * 100


if st_rate >= 150:
    performance = "KYa Baat Hai bhai ! 🏆"

elif st_rate >= 100:
    performance = "Average hai Thik Thik 👍"

else:
    performance = "Mehnat kar or 😔"

# Output
print("\n" + "=" * 45)
print("               SCORECARD")
print("=" * 45)

print(f"Total ovr     : {ovr}")
print(f"Total Balls     : {t_ball}")
print(f"Total Runs      : {t_run}")

print("-" * 45)

print(f"Fours Hit       : {fours}")
print(f"Sixes Hit       : {sixes}")
print(f"Dot Balls       : {d_ball}")

print("-" * 45)

print(f"Boundary Runs   : {boundry}")
print(f"Strike Rate     : {round(st_rate, 2)}")

print("-" * 45)

print(f"Performance     : {performance}")

print("=" * 45)


if t_run >= 100:
    print("Century Ban Gyi Bhai KI 🔥")

elif t_run >= 50:
    print("Half-Century Ho Gyi! 👏")

else:
    print("OR Mehnat KAro! 💪")

print("=" * 45)
print("          English Mai Dhanywaad 👐")
print("=" * 45)
```
