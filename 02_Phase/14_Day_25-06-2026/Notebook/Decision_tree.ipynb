# Decision Tree

```python
import numpy as np  
import pandas as pd 
import seaborn as sns 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score , confusion_matrix , f1_score , recall_score , precision_score , classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset
data = {
"Fever": [98,99, 100, 101, 102, 103,98,99,101,102],
"Cough": [0,0,1,1,1,1,0,1,1,1],
"BreathingProblem" : [0,0,0,1,1,1,0,0,1,1],
"COVID_Status":[
"Negative", "Negative", "Negative",
"Positive", "Positive", "Positive",
"Negative", "Negative","Positive", "Positive"
]
}
df = pd.DataFrame(data)

df['COVID_Status'] = df['COVID_Status'].map({'Negative' :0 , 'Positive':1})
df.astype(int)

print(df.columns)
df.head()
```

```python
col=['Fever', 'Cough', 'BreathingProblem']

X = df[col]
y = df['COVID_Status']

model = DecisionTreeClassifier()

model.fit(X,y)
print(df.head())

# User Input
fever = int(input("Enter  fever Degree : "))
cough = int(input("Enter cough (Yes: 1, No: 0): "))
breathingProblem = float(input("Enter breathingProblem (Yes: 1, No: 0) : "))

# prediction
result = model.predict([[fever, cough, breathingProblem]])[0]

prob = model.predict_proba([[fever, cough, breathingProblem]])

print(f"Rejected Probability: {round(prob[0][0] * 100, 2)}%")
print(f"Approved Probability: {round(prob[0][1] * 100, 2)}%")

# 6. Model Accuracy
y_pred = model.predict(X)
print(f"\nModel Accuracy: {round(accuracy_score(y, y_pred) * 100, 2)}%")
print("Prediciton :"  , "Positive" if result == 1 else "Negative")


# Confusion Matrix
    
cm = confusion_matrix(y, model.predict(X))
print("\n Confusion Matrix :")
print(cm)

# Extract TN, FP, FN, TP
TN, FP, FN, TP = cm.ravel()

print("\nTN =", TN)
print("FP =", FP)
print("FN =", FN)
print("TP =", TP)

# Manual Calculations
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)

print("\nManual Calculation")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)

# sklearn Calculations
print("\nUsing sklearn")
print("Accuracy :", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred, pos_label=1))
print("Recall   :", recall_score(y, y_pred, pos_label=1))
print("F1 Score :", f1_score(y, y_pred, pos_label=1))


```
