## 📊 Series — 1D Labeled Array

A **Series** is a one-dimensional labeled array capable of holding any data type.

**Components:**
- **Values** — the actual data
- **Index** — labels for each value

**Ways to Create a Series:**
- From a List
- From a Tuple
- From a Dictionary
- From a NumPy Array
- With Custom Index

```python
# Creating Series from a List (default integer index)
l = [90, 80, 70, 60, 50]
s2 = pd.Series(l)
print("Series from List:")
print(s2)
```

```python
# Creating Series from a Tuple
t = (90, 80, 70, 60, 50)
s3 = pd.Series(t)
print("Series from Tuple:")
print(s3)
```

```python
# Creating Series from a Dictionary
# Keys become the index labels automatically
d = {
    "name": "Aditya",
    "Status": "Duniya ka papa",
    "age": "ant se anant tak"
}
s1 = pd.Series(d)
print("Series from Dictionary:")
print(s1)
```

```python
# Accessing multiple elements from a dictionary-based Series
print("Accessing 'name' and 'age':")
print(s1[["name", "age"]])
```

```python
# Creating Series from NumPy Array with Custom String Index
n1 = np.array([10, 20, 30, 40, 50])

# Custom string index — accessing like a dictionary
s4 = pd.Series(n1, index=["a", "b", "c", "d", "e"])
print("Series with custom index:")
print(s4)
print("\nAccess s4['c']  :", s4["c"])              # 30
print("Access multiple  :", s4[["a", "c", "e"]].values)  # [10, 30, 50]
```

```python
# Creating Series with label index and name
lable = ["A", "B", "C", "D", "E"]
s1 = pd.Series(np.arange(10, 51, 10), index=lable, name="DAtA_OF DAtA")
print(s1)

# Various ways to slice
# print(s1[4])        # element at position 4
# print(s1[0:4])      # first 4 elements
# print(s1[::-1])     # reverse the series
# print(s1[0:2])      # first 2 elements
# print(s1[0::2])     # every alternate element
```

```python
# Slicing a Series
s2 = pd.Series([10, 20, 30, 40, 50])
print("Full Series:")
print(s2)
print("\nSlice [1:4]  :", s2[1:4].values)   # [20 30 40]
print("Reversed     :", s2[::-1].values)   # [50 40 30 20 10]
```

## 📋 Series Attributes

| Attribute | Returns | Description |
|-----------|---------|-------------|
| `.index` | Index object | All labels |
| `.values` | NumPy array | Raw data |
| `.dtype` | dtype | Data type |
| `.shape` | tuple | Dimensions |
| `.size` | int | Total elements |
| `.ndim` | int | Always 1 for Series |
| `.name` | str | Series name |

```python
# Exploring all Series attributes
s1 = pd.Series([10, 20, 30, 40, 50],
               index=["a", "b", "c", "d", "e"],
               name="Series")
print(s1)

print("\nindex   :", s1.index)    # Index labels
print("values  :", s1.values)   # NumPy array
print("dtype   :", s1.dtype)    # int64
print("shape   :", s1.shape)    # (5,)
print("size    :", s1.size)     # 5
print("ndim    :", s1.ndim)     # 1
print("name    :", s1.name)     # Series
```

```python
# Aggregation Functions on Series
lable = ["A", "B", "C", "D", "E"]
s1 = pd.Series(np.arange(10, 51, 10), index=lable, name="DAtA_OF DAtA")

print("Sum          :", sum(s1))                    # 150
print("Mean         :", s1.mean())                  # 30.0
print("Median       :", s1.median())                # 30.0
print("MAX          :", s1.max())                   # 50
print("MIN          :", s1.min())                   # 10
print("Count        :", s1.count())                 # 5
print("Std Dev      :", round(s1.std(), 2))         # 15.81
print("Variance     :", s1.var())                   # 250.0

# Mode example (repeated values)
# s = pd.Series([1, 1, 1, 1, 1, 1, 1, 1, 1])
# print(s.mode())
```

## 📋 DataFrame — 2D Labeled Table

A **DataFrame** is a two-dimensional, size-mutable, labeled data structure — like a spreadsheet or SQL table.

**Components:**
- **Data** — actual values
- **Rows** — records/entries
- **Columns** — features/attributes
- **Index** — row labels
- **Data Types** — dtype per column

**Ways to Create a DataFrame:**
- From Dictionary (most common)
- From List of Lists
- From List of Dictionaries
- From NumPy Array
- From CSV / Excel File

```python
# Creating DataFrame from a Dictionary
d = {
    "name": ["asif", "amit", "mohit", "himanshu"],
    "city": ["agra", "jaipur", "kota", "delhi"],
    "age":  [27, 23, 45, 34]
}
df = pd.DataFrame(d)
print(df)
```

## 👀 Viewing DataFrame Data

| Method | Description |
|--------|-------------|
| `head()` | First 5 rows |
| `tail()` | Last 5 rows |
| `sample()` | Random rows |
| `info()` | Column info & null counts |
| `describe()` | Statistical summary |

```python
# Viewing methods
print("=== head(2) ===")
print(df.head(2))

print("\n=== tail(2) ===")
print(df.tail(2))

print("\n=== info() ===")
df.info()

print("\n=== describe() ===")
print(df.describe())
```

## 📋 DataFrame Attributes

| Attribute | Description |
|-----------|-------------|
| `.index` | Row labels |
| `.columns` | Column names |
| `.dtypes` | Data types per column |
| `.shape` | `(rows, cols)` tuple |
| `.size` | Total number of elements |
| `.ndim` | Number of dimensions (always 2) |
| `.values` | NumPy array representation |
| `.axes` | List of axes |

```python
# DataFrame Attributes
print("index   :", df.index)    # RangeIndex(start=0, stop=4, step=1)
print("columns :", df.columns)  # Index(['name', 'city', 'age'], dtype='object')
print("dtypes  :\n", df.dtypes)
print("shape   :", df.shape)    # (4, 3)
print("size    :", df.size)     # 12
print("ndim    :", df.ndim)     # 2
print("values  :\n", df.values)
```

```python
# Column Selection
print("=== Single Column (Series) ===")
print(df["name"])

print("\n=== Multiple Columns (DataFrame) ===")
print(df[["name", "city"]])

# Create Series from a column
s1 = pd.Series(df["city"])
print("\n=== City as Series ===")
print(s1)
```

```python
# Row Selection — iloc (integer position based)
print("=== iloc[0] — First Row ===")
print(df.iloc[0])

print("\n=== iloc[0:2] — First 2 Rows ===")
print(df.iloc[0:2])

# Row Selection — loc (label based)
print("\n=== loc[0, 'name'] — Specific Cell ===")
print(df.loc[0, "name"])

print("\n=== loc[0:2, ['name','city']] — Rows 0–2, selected columns ===")
print(df.loc[0:2, ["name", "city"]])
```

```python
# Reading from a CSV file — Real-world data
# Uncomment the line below to read your actual CSV file
# file = pd.read_csv("wed17.csv")

# file.head()          # First 5 rows
# file.info()          # Schema info
# file.describe()      # Statistical summary
# file.shape           # (rows, cols)
# file.columns         # Column names

# Example with a simple inline dictionary simulating CSV data:
sample_csv_data = {
    "student_id": [101, 102, 103, 104],
    "name": ["Aditya", "Riya", "Mohit", "Sara"],
    "marks": [88, 92, 75, 96],
    "grade": ["A", "A+", "B", "A+"]
}
sample_df = pd.DataFrame(sample_csv_data)
print("Sample Student Data:")
print(sample_df)
print("\ndescribe():")
print(sample_df.describe())
```

```python
# Filtering Data — Boolean Indexing
print("Students with marks > 80:")
print(sample_df[sample_df["marks"] > 80])

print("\nStudents with grade A+:")
print(sample_df[sample_df["grade"] == "A+"])
```

```python
# Sorting Data
print("=== Sorted by marks (Ascending) ===")
print(sample_df.sort_values("marks"))

print("\n=== Sorted by marks (Descending) ===")
print(sample_df.sort_values("marks", ascending=False))
```

```python
# Aggregation on DataFrame
print("Sum of marks       :", sample_df["marks"].sum())
print("Mean marks         :", sample_df["marks"].mean())
print("Median marks       :", sample_df["marks"].median())
print("Max marks          :", sample_df["marks"].max())
print("Min marks          :", sample_df["marks"].min())
print("Count              :", sample_df["marks"].count())
```

```python
# Data Transformation — Add, rename, drop columns
df_copy = sample_df.copy()

# Add a new column
df_copy["passed"] = df_copy["marks"] >= 80
print("DataFrame with 'passed' column:")
print(df_copy)

# Rename a column
df_copy.rename(columns={"name": "Student Name"}, inplace=True)

# Drop a column
df_copy.drop("grade", axis=1, inplace=True)

print("\nAfter rename and drop:")
print(df_copy)
```
