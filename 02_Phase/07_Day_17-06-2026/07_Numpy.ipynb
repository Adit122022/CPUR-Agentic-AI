# Indexing  of array (numpy)

```python
import numpy as np
```

```python
a1 = np.arange(1,11,2)
print(a1)
print(a1[2:])
print(a1[5:2:-1])
```

```python
r = (np.random.rand(4,4)*100).astype(int)
print(r)
print("=="*50)
# print("[1:3,0:2]  :",r[1:3,0:2])
# print(r[-3:-1,-1])
print(r[:,::2])
```
