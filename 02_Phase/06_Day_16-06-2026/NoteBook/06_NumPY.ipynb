## ⚡ Why Do We Use NumPy?

Python lists are good for small data. But for thousands/millions of records and math calculations, lists become slower.

**NumPy solves these problems by:**
- ✅ Faster execution
- ✅ Less memory consumption
- ✅ Easy mathematical calculations
- ✅ Multi-dimensional arrays
- ✅ Better performance

## 🔢 NumPy vs Python Lists

| Feature | Python List | NumPy Array |
|---------|-------------|-------------|
| Speed | Slow | **Fast** |
| Memory Usage | More | **Less** |
| Mathematical Operations | Difficult | **Easy** |
| Data Type | Mixed Allowed | Same Type Preferred |
| Multi-Dimensional Support | Limited | **Excellent** |
| Scientific Computing | Not Suitable | **Designed For It** |

````python
# ❌ Python List — Manual loop needed for element-wise addition
a = [1, 2, 3]
b = [4, 5, 6]
result = []
for i in range(len(a)):
    result.append(a[i] + b[i])
print("List result:", result)   # [5, 7, 9]

# ✅ NumPy — One clean vectorized operation!
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a + b
print("NumPy result:", result)  # [5 7 9]
````

````python
# ❌ Python List — multiply each element by 2
numbers = [1, 2, 3, 4, 5]
result = []
for num in numbers:
    result.append(num * 2)
print("List:", result)   # [2, 4, 6, 8, 10]

# ✅ NumPy — Broadcasting! Scalar applied to all elements
numbers = np.array([1, 2, 3, 4, 5])
result = numbers * 2
print("NumPy:", result)  # [ 2  4  6  8 10]
````

## 🧮 Creating Arrays

NumPy supports:
- **1D Array** — single row of elements
- **2D Array** — matrix (rows × columns)
- **3D Array** — tensor (blocks × rows × columns)

**Key Attributes:**
- `ndim` → Number of Dimensions
- `shape` → Array Shape (tuple)
- `size` → Total number of Elements
- `dtype` → Data Type of elements
- `itemsize` → Bytes per element

````python
# 1D Array
d1 = np.array([10, 20, 30, 40, 50])

print("Array    :", d1)
print("ndim     :", d1.ndim)      # 1
print("shape    :", d1.shape)     # (5,)
print("size     :", d1.size)      # 5
print("dtype    :", d1.dtype)     # int64
print("itemsize :", d1.itemsize)  # 8 bytes
````

````python
# 2D Array — Matrix
d2 = np.array([
    [1, 2, 3],
    [8, 9, 0],
    [3, 4, 5]
])

print("2D Array:\n", d2)
print("ndim  :", d2.ndim)     # 2
print("shape :", d2.shape)    # (3, 3)
print("size  :", d2.size)     # 9
print("dtype :", d2.dtype)    # int64
````

````python
# 3D Array — Tensor (3 blocks, each 3x3)
d3 = np.array([
    [[1, 2, 3], [8, 9, 0], [3, 4, 5]],
    [[1, 2, 3], [8, 9, 0], [3, 4, 5]],
    [[1, 2, 3], [8, 9, 0], [3, 4, 5]],
])

print("3D Array:\n", d3)
print("ndim  :", d3.ndim)     # 3
print("shape :", d3.shape)    # (3, 3, 3)
````

## ➕ Mathematical Calculations

NumPy performs element-wise operations and broadcasting automatically.

**Supported Operations:**
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division
- `//` Floor Division
- `%` Modulus
- `**` Power

````python
d1 = np.array([10, 20, 30])
d2 = np.array([11, 12, 13])

print("d1 + d2 :", d1 + d2)   # [21 32 43] — element-wise
print("d1 + 5  :", d1 + 5)    # [15 25 35] — scalar broadcasting
print("d2 - 5  :", d2 - 5)    # [ 6  7  8] — scalar subtraction
print("d1 * d2 :", d1 * d2)   # [110 240 390]
print("d1 ** 2 :", d1 ** 2)   # [100 400 900]
````

## 🏗️ Creating Arrays Using Different Functions

| Function | Description | Syntax |
|----------|-------------|--------|
| `array()` | Create from list/tuple | `np.array([1,2,3])` |
| `zeros()` | All zeros | `np.zeros(shape, dtype=float)` |
| `ones()` | All ones | `np.ones(shape)` |
| `arange()` | Like Python range | `np.arange(start, stop, step)` |
| `linspace()` | Evenly spaced | `np.linspace(start, stop, num)` |
| `empty()` | Uninitialized | `np.empty(shape)` |
| `full()` | Fill with value | `np.full(shape, value)` |
| `eye()` | Identity matrix | `np.eye(n)` |
| `random.rand()` | Random [0,1) | `np.random.rand(shape)` |
| `random.randint()` | Random integers | `np.random.randint(low, high, shape)` |
| `reshape()` | Change shape | `arr.reshape(new_shape)` |

````python
# zeros() — Create arrays filled with zeros
z1 = np.zeros(5, dtype=str)           # 1D string array of empty strings
z2 = np.zeros((3, 4))                 # 3x4 matrix of 0.0 (float)
z3 = np.zeros((3, 4, 2), dtype=bool)  # 3D bool array of False

print("z1 (1D str)  :", z1)
print("z2 (3x4 float):\n", z2)
print("z3 shape:", z3.shape, "| dtype:", z3.dtype)
````

## 🔪 Array Indexing & Slicing

### 1D Array Slicing Cheat Sheet

| Operation | Syntax | Description |
|-----------|--------|-------------|
| Entire Array | `arr[:]` | All elements |
| First Element | `arr[0]` | Index 0 |
| Last Element | `arr[-1]` | Last index |
| First 3 | `arr[:3]` | Indices 0–2 |
| Last 3 | `arr[-3:]` | Last 3 |
| Every 2nd | `arr[::2]` | Step 2 |
| Reverse | `arr[::-1]` | Reversed |

### 2D — Row Operations

| Operation | Syntax |
|-----------|--------|
| First Row | `arr[0]` |
| Last Row | `arr[-1]` |
| Row 1, all cols | `arr[1, :]` |
| All Rows | `arr[:, :]` |

### 2D — Column Operations

| Operation | Syntax |
|-----------|--------|
| First Column | `arr[:, 0]` |
| Last Column | `arr[:, -1]` |
| First Two Cols | `arr[:, 0:2]` |
| Alternate Cols | `arr[:, ::2]` |

### 3D Array Selection

| Operation | Syntax |
|-----------|--------|
| First Block | `arr[0]` |
| Specific Element | `arr[0, 1, 2]` |
| Entire First Block | `arr[0, :, :]` |

````python
# 2D Indexing Examples — Random 3x4 array
d3 = (np.random.rand(3, 4) * 100).astype(int)
print("Array:\n", d3)
print("="*40)

# Access specific element at row 2, column -3 (third from last)
print("d3[2, -3]  :", d3[2, -3])

# Access entire column -3 (third from last column)
print("d3[:, -3]  :", d3[:, -3])
````

````python
# 2D slicing with alternate columns
r = (np.random.rand(4, 4) * 100).astype(int)
print("Original 4x4 Array:\n", r)
print("="*50)

# Select alternate columns (columns 0 and 2)
print("Alternate Columns arr[:, ::2]:\n", r[:, ::2])
````

## 📊 Aggregation Functions

| Function | Description |
|----------|-------------|
| `np.sum()` | Sum of all elements |
| `np.mean()` | Arithmetic mean |
| `np.median()` | Median value |
| `np.min()` | Minimum value |
| `np.max()` | Maximum value |
| `np.std()` | Standard deviation |
| `np.var()` | Variance |

````python
# Aggregation Functions
d1 = np.array([10, 11, 13, 12, 10])

print("Array       :", d1)
print("sum()       :", np.sum(d1))     # 56
print("median()    :", np.median(d1))  # 11.0

# Larger array with .attribute notation
a = np.array([10, 20, 30, 40, 50, 60, 80])
print("\nArray       :", a)
print("Mean        :", a.mean())         # 41.43
print("Std Dev     :", a.std())          # 22.31
print("Variance    :", a.var())          # 497.96
print("Min         :", a.min())          # 10
print("Max         :", a.max())          # 80
````
