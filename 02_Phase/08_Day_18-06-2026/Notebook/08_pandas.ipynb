```python
# Warm-up Exercise: Simple function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False 
    return True

print("Is 7 prime?:", is_prime(7))
```

```python
df =pd.read_csv("../DataSet/train.csv")
df
```

```python
# Displaying unique values and counts for columns
print("Unique gender categories in dataset:")
print(df["Sex"].unique())

print("\nNumber of unique values in each column:")
print(df.nunique())

print("\nCount of non-null values in each column:")
print(df.count())

print("\nNumber of unique names in 'Name' column:")
print(df["Name"].nunique())

print("\nTotal non-null names count:")
print(df["Name"].count())
```
