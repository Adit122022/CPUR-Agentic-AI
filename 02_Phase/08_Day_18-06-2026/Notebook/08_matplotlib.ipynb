```python
# Creating a simple Stackplot
x = [1, 2, 3, 4, 5]
y = [50, 40, 50, 80, 10]

plt.stackplot(x, y, color="black")
plt.title("Simple Area / Stackplot Graph")
plt.xlabel("X-axis (Sequence)")
plt.ylabel("Y-axis (Values)", color="red")
plt.grid(True, linestyle=":", alpha=0.6)
plt.show()
```

```python
# Creating a Horizontal Bar Graph with custom colors
x = [1, 2, 3, 4, 5]
y = [50, 40, 50, 80, 10]

plt.barh(x, y, color=["red", "green", "blue", "orange", "purple"])
plt.title("Simple Horizontal Bar Graph")
plt.xlabel("Marks (Y-axis)", color="blue")
plt.ylabel("Student ID (X-axis)", color="red")
plt.show()
```

```python
# Plotting multiple lines on the same graph with a Legend
x1 = [1, 2, 2, 3]
y = [80, 80, 0, 0]
x2 = [1, 1, 3, 3]
y1 = [0, 40, 40, 80]

plt.plot(x1, y, color="green", label="Python")
plt.plot(x2, y1, color="black", label="Java")
plt.legend()

plt.title("Multi-line Comparison Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis", color="red")
plt.grid(True)
plt.show()
```

```python
# Plotting survival rate using gender submission file
df = pd.read_csv("../DataSet/gender_submission.csv")
data = df.head(20)

plt.bar(data["PassengerId"], data['Survived'], color="green")
plt.title("Titanic Passenger Survival Rate (First 20 Records)")
plt.xlabel("Passenger ID", color="green")
plt.ylabel("Survived (1) / Died (0)", color="red")
plt.xticks(data["PassengerId"], rotation=90)
plt.show()
```

# Create PLOTS using dataFrame ( -me)

```python
# Creating a customized Line Graph with grid and markers
x = [1, 2, 3, 4, 5]
y = [50, 40, 50, 80, 10]

plt.plot(x, y, color="green", linestyle="--", marker="o")
plt.grid(True, linestyle="--", alpha=0.5)

plt.title("Simple Line Graph with Markers")
plt.xlabel("X-axis (Independent variable)")
plt.ylabel("Y-axis (Dependent variable)", color="red")
plt.show()
```

```python
# Comparison of Python vs Java progress over timeline
x = [1, 2, 3, 4, 5]
y = [50, 40, 50, 80, 10]
y1 = [44, 88, 22, 65, 35]

plt.plot(x, y, color="green", linestyle="--", marker="o", label="Python")
plt.plot(x, y1, color="red", linestyle="--", marker="o", label="Java")
plt.legend()

plt.title("Python vs Java Timeline Comparison")
plt.xlabel("Timeline (Days)")
plt.ylabel("Performance / Value", color="red")
plt.grid(True)
plt.show()
```
