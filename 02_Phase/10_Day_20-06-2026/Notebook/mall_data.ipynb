```python
gender_counts = df['Genre'].value_counts()
print("\nGender Counts :")
print(gender_counts)
# print("\n Female mean :",gender_counts['Female'].mean())
# print("\n Male mean :",gender_counts['Male'].mean())
```

```python
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', edgecolor='black')
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.pie(gender_counts, labels =["female" , "male"],autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], alpha=0.7,  s=60)
plt.title('Annual  vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

```python
# Age Vs Spending score
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Spending Score (1-100)'],  alpha=0.7, s=60)
plt.title('Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Annual Income (k$)'],  alpha=0.7, s=60)
plt.title('Age vs Anual  Income')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
