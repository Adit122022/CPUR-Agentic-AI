### Step 10: Outlier Detection (using Matplotlib Boxplots)


```python
# Visualize boxplots of all 4 numerical columns using Matplotlib
num_cols = ["raisedhands", "VisITedResources", "AnnouncementsView", "Discussion"]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for idx, col in enumerate(num_cols):
    r, c = idx // 2, idx % 2
    axes[r, c].boxplot(df[col], vert=False, patch_artist=True, boxprops=dict(facecolor="#5c9eb2"))
    axes[r, c].set_title(f"Boxplot of {col}")
    axes[r, c].set_xlabel(col)
    axes[r, c].grid(True, linestyle="--", alpha=0.5)
    
plt.suptitle("Outlier Check for Academic Activity Indicators", fontsize=16)
plt.tight_layout()
plt.show()

# No significant outliers lie outside the 0-100 range since these are behavioral counts.
```

### Step 11: Perform Univariate Analysis


```python
# Distribution of Target Variable Class Level (L=Low, M=Medium, H=High)
class_counts = df["Class"].value_counts().reindex(["L", "M", "H"])

plt.figure(figsize=(6, 4))
plt.bar(class_counts.index, class_counts.values, color=["#e74c3c", "#f1c40f", "#2ecc71"], edgecolor="black", width=0.6)
plt.title("Distribution of Student Performance Class")
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()

print("Performance Class Percentages:")
print(df["Class"].value_counts(normalize=True) * 100)
```

```python
# Distribution of Gender
gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color=["#3498db", "#e74c3c"], edgecolor="black", width=0.6)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()

print("Gender Counts:")
print(df["gender"].value_counts())
```

```python
# Distribution of Nationality
nat_counts = df["NationalITy"].value_counts()

plt.figure(figsize=(10, 6))
plt.barh(nat_counts.index[::-1], nat_counts.values[::-1], color="#34495e", edgecolor="black")
plt.title("Distribution of Student Nationalities")
plt.xlabel("Student Count")
plt.ylabel("Nationality")
plt.grid(True, axis="x", linestyle="--", alpha=0.5)
plt.show()
```

### Step 12: Perform Bivariate Analysis (Standard Aggregations)


```python
# Correcting user's raw bar plots by grouping/boxplotting values to show correct relations.

# 1. Performance Class vs Raised Hands (using boxplot by category)
df.boxplot(column="raisedhands", by="Class", grid=True, figsize=(7, 5))
plt.title("Raised Hands Distribution by Performance Class")
plt.suptitle("") # Remove default pandas title
plt.xlabel("Performance Class")
plt.ylabel("Number of Raised Hands")
plt.show()
```

```python
# 2. Gender vs Raised Hands
df.boxplot(column="raisedhands", by="gender", grid=True, figsize=(7, 5))
plt.title("Raised Hands Distribution by Gender")
plt.suptitle("")
plt.xlabel("Gender")
plt.ylabel("Number of Raised Hands")
plt.show()
```

```python
# 3. Student Absences vs Performance Class (using pandas bar plots)
crosstab_absence = pd.crosstab(df["StudentAbsenceDays"], df["Class"])
print("Absence Days vs Class Level:")
display(crosstab_absence)

crosstab_absence.T.plot(kind="bar", color=["#e74c3c", "#2ecc71"], edgecolor="black", width=0.7)
plt.title("Student Performance Class vs Absence Days")
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.xticks(rotation=0)
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
# 4. Parent satisfaction vs Performance Class
crosstab_sat = pd.crosstab(df["ParentschoolSatisfaction"], df["Class"])
crosstab_sat.T.plot(kind="bar", color=["#95a5a6", "#27ae60"], edgecolor="black", width=0.7)
plt.title("Performance Class by Parent School Satisfaction")
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.xticks(rotation=0)
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()
```

### Step 13: Correlation Analysis


```python
# Compute correlation matrix for academic indicators
corr_matrix = df[num_cols].corr()

# Generate Heatmap in Matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap="Blues", interpolation="none", vmin=-1, vmax=1)
plt.colorbar(label="Correlation")

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# Annotate correlation text
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha="center", va="center", color="black", fontweight="bold")

plt.title("Correlation Map: Student Activities")
plt.tight_layout()
plt.show()
```

### Step 14: Feature Engineering


```python
# We can create a composite Engagement Score by summing all academic behavior scores
df["EngagementScore"] = df["raisedhands"] + df["VisITedResources"] + df["AnnouncementsView"] + df["Discussion"]

print(df[["raisedhands", "VisITedResources", "AnnouncementsView", "Discussion", "EngagementScore"]].head())

# Plot Engagement Score against the Performance Class
df.boxplot(column="EngagementScore", by="Class", grid=True, figsize=(8, 5))
plt.title("Overall Student Engagement Score by Performance Class")
plt.suptitle("")
plt.xlabel("Performance Class")
plt.ylabel("Engagement Score (Sum of Activities)")
plt.show()
```

### Step 15: Draw Conclusions and Insights
1. **Engagement strongly correlates with grades**: Students who raise hands more, visit resources, view announcements, and participate in discussions belong to the High ('H') academic category, as demonstrated by the Engagement Score boxplot.
2. **Attendance matters**: Students with fewer absence days ('Under-7') dominate the High and Medium performance categories, whereas students with high absences ('Above-7') mostly fall into the Low ('L') performance category.
3. **Parents' School Satisfaction**: Parent involvement and satisfaction are positively associated with student outcomes. High-performing students tend to have parents who express higher satisfaction.
4. **Gender & Participation**: Female students on average raised hands slightly more than male students, as shown in the boxplot.
