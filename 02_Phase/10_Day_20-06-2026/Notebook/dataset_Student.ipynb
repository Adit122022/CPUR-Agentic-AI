### Step 7: Check Duplicate Records


```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate records: {duplicate_count}")

# Let's inspect the duplicate rows
if duplicate_count > 0:
    display(df[df.duplicated(keep=False)].head(6))
```

### Step 8: Remove Duplicates


```python
# Fix the user's syntax error: df.drop_duplicates(all_duplicates, inplace=True)
# Correct syntax is simply calling drop_duplicates()
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicates successfully removed!")
print("Cleaned shape:", df.shape)
```

### Step 10: Detect Outliers (IQR Method)


```python
# Let's visualize boxplots of all 4 numerical columns to check for outliers
num_cols = ["raisedhands", "VisITedResources", "AnnouncementsView", "Discussion"]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for idx, col in enumerate(num_cols):
    r, c = idx // 2, idx % 2
    sns.boxplot(x=df[col], ax=axes[r, c], color="skyblue")
    axes[r, c].set_title(f"Boxplot of {col}")
    
plt.suptitle("Outlier Check for Academic Activity Indicators", fontsize=16)
plt.tight_layout()
plt.show()

# No significant outliers lie outside the 0-100 range since these are behavioral counts.
```

### Step 11: Perform Univariate Analysis


```python
# Distribution of Target Variable (Class Level: L=Low, M=Medium, H=High)
plt.figure(figsize=(6, 4))
ax = sns.countplot(x="Class", data=df, order=["L", "M", "H"], palette="viridis")
plt.title("Distribution of Student Performance Class")
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.show()

print("Performance Class Percentages:")
print(df["Class"].value_counts(normalize=True) * 100)
```

```python
# Distribution of Gender
plt.figure(figsize=(6, 4))
sns.countplot(x="gender", data=df, palette="pastel")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

print("Gender Counts:")
print(df["gender"].value_counts())
```

```python
# Distribution of Nationality
plt.figure(figsize=(12, 5))
sns.countplot(y="NationalITy", data=df, order=df["NationalITy"].value_counts().index, palette="mako")
plt.title("Distribution of Student Nationalities")
plt.xlabel("Student Count")
plt.ylabel("Nationality")
plt.show()
```

### Step 12: Perform Bivariate Analysis


```python
# Correcting user's raw bar plots: 
# Instead of plt.bar(df['Class'], df['raisedhands']), we use Seaborn's barplot/boxplot to show aggregation.

# 1. Performance Class vs Raised Hands (Categorical vs Numerical)
plt.figure(figsize=(7, 5))
sns.boxplot(x="Class", y="raisedhands", data=df, order=["L", "M", "H"], palette="coolwarm")
plt.title("Raised Hands Distribution by Performance Class")
plt.xlabel("Performance Class")
plt.ylabel("Number of Raised Hands")
plt.show()
```

```python
# 2. Gender vs Raised Hands
plt.figure(figsize=(7, 5))
sns.boxplot(x="gender", y="raisedhands", data=df, palette="pastel")
plt.title("Raised Hands Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Raised Hands")
plt.show()
```

```python
# 3. Student Absences vs Performance Class (Categorical vs Categorical)
crosstab_absence = pd.crosstab(df["StudentAbsenceDays"], df["Class"])
print("Absence Days vs Class Level:")
display(crosstab_absence)

plt.figure(figsize=(7, 5))
sns.countplot(x="Class", hue="StudentAbsenceDays", data=df, order=["L", "M", "H"], palette="muted")
plt.title("Student Performance Class vs Absence Days")
# Under-7 means < 7 days absence, Above-7 means >= 7 days absence
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.show()
```

```python
# 4. Parent satisfaction vs Performance Class
plt.figure(figsize=(7, 5))
sns.countplot(x="Class", hue="ParentschoolSatisfaction", data=df, order=["L", "M", "H"], palette="Set1")
plt.title("Performance Class by Parent School Satisfaction")
plt.xlabel("Performance Class")
plt.ylabel("Student Count")
plt.show()
```

### Step 13: Correlation Analysis


```python
# Compute correlation heatmap for academic activity indicators
plt.figure(figsize=(8, 6))
corr_matrix = df[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="Blues", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Student Academic Activity Indicators")
plt.show()
```

### Step 14: Feature Engineering


```python
# We can create a composite Engagement Score by summing all academic behavior scores
df["EngagementScore"] = df["raisedhands"] + df["VisITedResources"] + df["AnnouncementsView"] + df["Discussion"]

print(df[["raisedhands", "VisITedResources", "AnnouncementsView", "Discussion", "EngagementScore"]].head())

# Let's plot Engagement Score against the Performance Class
plt.figure(figsize=(8, 5))
sns.boxplot(x="Class", y="EngagementScore", data=df, order=["L", "M", "H"], palette="rainbow")
plt.title("Overall Student Engagement Score by Performance Class")
plt.xlabel("Performance Class")
plt.ylabel("Engagement Score (Sum of Activities)")
plt.show()
```

### Step 15: Draw Conclusions and Insights
1. **Engagement strongly correlates with grades**: Students who raise hands more, visit resources, view announcements, and participate in discussions belong to the High ('H') academic category, as demonstrated by the Engagement Score boxplot.
2. **Attendance matters**: Students with fewer absence days ('Under-7') dominate the High and Medium performance categories, whereas students with high absences ('Above-7') mostly fall into the Low ('L') performance category.
3. **Parents' School Satisfaction**: Parent involvement and satisfaction are positively associated with student outcomes. High-performing students tend to have parents who respond to school surveys and express higher satisfaction.
4. **Gender & Participation**: Female students on average raised hands slightly more than male students, as shown in the boxplot.
