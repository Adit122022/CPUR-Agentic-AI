### Missing Value Imputation
- Fill missing `Age` with the median age.
- Fill missing `Salary` with the mean salary.
- Fill missing `City` with the mode (most frequent) city.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df.isnull().sum()
```

```python
df["Salary"] = df["Salary"].fillna(round(df["Salary"].mean()))
df.isnull().sum()
```

```python
df["City"] = df["City"].fillna(df["City"].mode()[0])
df.isnull().sum()
```

### Outlier Detection using the IQR Method
1. Calculate First Quartile ($Q_1$) and Third Quartile ($Q_3$).
2. Calculate Interquartile Range ($IQR = Q_3 - Q_1$).
3. Define boundaries: `lower_limit = Q1 - 1.5 * IQR`, `upper_limit = Q3 + 1.5 * IQR`.
4. Filter out data points lying outside these boundaries.

```python
# Calculate Q1 and Q3
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_limit}, Upper Bound: {upper_limit}")

# Filter Outliers
df_clean = df[(df["Age"] >= lower_limit) & (df["Age"] <= upper_limit)]
print("Original shape:", df.shape)
print("Cleaned shape:", df_clean.shape)
df_clean
```
