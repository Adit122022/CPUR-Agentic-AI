# fill the is na or null value 

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
df
```

```python
# salary 
df["Salary"]=df["Salary"].fillna(round(df["Salary"].mean()))
df
```

```python
# City 
df["City"]=df["City"].fillna(df["City"].mode()[0])
df["City"].mode()[0]
```
