```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Part 1: Pie Charts

```python
# Basic Pie Chart with Title
data = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]
plt.pie(data, labels=labels)
plt.title("Student Data")
plt.show()
```

```python
# Pie Chart with Percentages
data = [40, 30, 20, 10]
labels = ["Python", "Java", "C++", "JS"]
plt.pie(data, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
# Pie Chart with Exploded Slices
data = [40, 30, 20, 10]
labels = ["Python", "Java", "C++", "JS"]
explode = [0.2, 0.1, 0, 0]
plt.pie(data, labels=labels, explode=explode, autopct="%1.1f%%")
plt.show()
```

```python
# Pie Chart with Shadow Effect
data = [40, 30, 20, 10]
labels = ["Python", "Java", "C++", "JS"]
explode = [0.2, 0.1, 0, 0]
plt.pie(data, labels=labels, explode=explode, shadow=True, autopct="%1.1f%%")
plt.show()
```

```python
# Starting Angle Rotation
data = [20, 20, 20, 20, 20]
labels = ["Python", "Java", "C++", "JS", "Bharat"]
plt.pie(data, labels=labels, shadow=True, startangle=90, autopct="%1.1f%%")
plt.show()
```

```python
# Pie Chart Customization (Borders & Edges)
data = [20, 20, 20, 20, 20]
labels = ["Python", "Java", "C++", "JS", "Bharat"]
plt.pie(
    data, 
    labels=labels,
    wedgeprops={
        'edgecolor': "black",
        'linewidth': 2
    },
    shadow=True,
    startangle=180,
    autopct="%1.1f%%"
)
plt.show()
```

```python
# Creating Subplots (Multiple Charts in One Figure)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
data1 = [50, 30, 20]
labels1 = ["Python", "Java", "C++"]
plt.pie(data1, labels=labels1)
plt.title("Chart 1")

plt.subplot(2, 2, 2)
data2 = [40, 35, 25]
labels2 = ["HTML", "CSS", "JS"]
plt.pie(data2, labels=labels2)
plt.title("Chart 2")

plt.subplot(2, 2, 3)
data3 = [10, 20, 70]
labels3 = ["Bharat", "Ayan", "Papa"]
plt.pie(data3, labels=labels3)
plt.title("Tabhai Bhai")

plt.tight_layout()
plt.show()
```

```python
# Saving the plot
data = [20, 20, 20, 20, 20]
labels = ["A", "B", "C", "D", "E"]
plt.pie(data, labels=labels, shadow=True, startangle=0, autopct="%1.1f%%")
plt.savefig("pie_chart.png")
plt.show()
```

## Part 2: Histograms

```python
# Basic Histogram on Random Data
x = 4 + np.random.normal(0, 1.5, 200)
plt.hist(x, bins=8, linewidth=0.5, edgecolor="white")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("../../DataSet/test.csv")
df.head()
```

```python
# Plotting Age and Sex from the Titanic dataset
df["Age"] = df["Age"].fillna(df["Age"].median())

plt.figure(figsize=(10, 6))
plt.hist(df["Sex"], bins=10, label="Sex", alpha=0.7, edgecolor="black")
plt.hist(df["Age"], bins=10, label="Age", alpha=0.7, edgecolor="black")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sex and Age Distributions")
plt.show()
```

## Part 3: Scatter Plots

```python
# Scatter Plot of PassengerId vs Age
plt.scatter(df["PassengerId"], df["Age"], c="red", alpha=0.6, edgecolor="black")
plt.xlabel("Passenger ID")
plt.ylabel("Age")
plt.title("Passenger ID vs Age Scatter Plot")
plt.show()
```

```python
# Custom Scatter Plot with Color and Size Mappings
plt.style.use('default') # Reset style

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots(figsize=(6, 5))
scatter = ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100, cmap='viridis')
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.colorbar(scatter, ax=ax, label="Color Value Scale")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Customized Scatter Plot")
plt.show()
```
