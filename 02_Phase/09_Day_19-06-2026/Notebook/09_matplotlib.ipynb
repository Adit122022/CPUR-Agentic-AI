```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

# adding title

```python
data =[40,30,20,10]
labels =["A", "B", "C", "D"]
plt.pie(data, labels= labels)
plt.title("Student Data")
plt.show()
# title() -> graph title


```

# Showing Percentage

```python
data =[40,30,20,10]
labels =["Python", "Java", "C++", "JS"]
plt.pie(
data, labels =labels,
autopct="%1.1f%%"
)
plt.show()
```

# explod in pie chart

```python
data =[40,30,20,10]
labels =["Python", "Java", "C++", "JS"]
explode = [0.2,0.1,0,0]
plt.pie(
data, 
labels =labels,
explode=explode,
autopct="%1.1f%%"
)
plt.show()
```

```python
# Shadow Effect 
data =[40,30,20,10]
labels =["Python", "Java", "C++", "JS"]
explode = [0.2,0.1,0,0]
plt.pie(
data, 
labels =labels,
explode=explode,
shadow=True,
autopct="%1.1f%%"
)
plt.show()
```

```python
# Starting  Angle 
data =[20,20,20,20,20]
labels =["Python", "Java", "C++", "JS","Bharat"]
# explode = [0.2,0.1,0,0]
plt.pie(
data, 
labels =labels,
# explode=explode,
shadow=True,
startangle=0,
autopct="%1.1f%%"
)
plt.show()
```

# pie chart border

```python
data =[20,20,20,20,20]
labels =["Python", "Java", "C++", "JS","Bharat"]
# explode = [0.2,0.1,0,0]
plt.pie(
data, 
labels =labels,
wedgeprops={
    'edgecolor':"black",
    'linewidth':5
},
shadow=True,
startangle=180,
# autopct: 'str | Callable[[float], str] | None' = None,
#     pctdistance: 'float' = 0.6,
#     shadow: 'bool' = False,
#     labeldistance: 'float | None' = 1.1,
#     startangle: 'float' = 0,
#     radius: 'float' = 1,
#     counterclock: 'bool' = True,
#     wedgeprops: 'dict[str, Any] | None' = None,
#     textprops: 'dict[str, Any] | None' = None,
#     center: 'tuple[float, float]' = (0, 0),
#     frame: 'bool' = False,
#     rotatelabels: 'bool' = False,
#     normalize: 'bool' = True,
#     hatch: 'str | Sequence[str] | None' = None,
#     data=None,
autopct="%1.1f%%"
)
plt.show()
```

```python
# Multiple Pie Charts
plt.subplot(2,2,1)
data1 =[50,30,20]
labels1 = ["Python", "Java", "C++"]
plt.pie(data1, labels=labels1)
plt.title("Chart 1")
plt.subplot(2,2,2)
data2 =[40,35,25]
labels2 = ["HTML", "CSS","JS"]
plt.pie(data2, labels=labels2)
plt.title("Chart 2")
plt.subplot(2,2,3)
data2 =[10,20,70]
labels2 = ["Bharat", "Ayan","Papa"]
plt.pie(data2, labels=labels2)
plt.title("Tabhai Bhai ")
plt.show()
# subplot(rows, cols, index)
```

```python
# saving pi chart

data =[20,20,20,20,20]
labels =["A", "B", "C" ,"D","E"]
plt.pie(
data, 
labels =labels,
shadow=True,
startangle=0,
autopct="%1.1f%%"
)
plt.savefig("badmosh.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))
print(colors)

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100 )

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
